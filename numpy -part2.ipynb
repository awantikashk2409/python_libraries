```python
# copy and view
a = np.array([2,3,4,5])
x = a.copy()
x[0] = 42
print(a)
print(x)
```

```python
a = np.array([2,3,4,5])
x = a.view()
x[0] = 42
print(a)
print(x)
```

```python
cop is a new array and view is just a view of the original array.
```
