```python
#2 using dictionary
df_dict = { 'Year' : [1990,1991,1992,1993],
           'country' : ['america','brazil','spain','germany'],
           'winner' : [ 'germany','spain','brazil','america'],
           'goalscored' : [115,112,113,114]
         }
pd.DataFrame(df_dict)
```

```python
#3 using series
df_dict_udindex = {'Winner' : pd.Series(['America','Brazil','Spain','Germany']),
                   'GoalScored' : pd.Series([115,116,117,118,119]),
                    }
pd.DataFrame(df_dict_udindex)
```
