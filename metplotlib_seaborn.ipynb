```python
import matplotlib.pyplot as plt
```

```python
week_day = [1,2,3,4,5,6,7]
temp = [50,51,49,45,47,32,50]
```

```python
plt.plot(week_day,temp, color = "green",linestyle = "--", marker = 'v') # the plot should be done only on 2 dimensions
plt.xlabel("weekday")
plt.ylabel("temperature")
plt.title("weekday vs temp")
```

```python
plt.plot() in Matplotlib is always 2-dimensional:

The first argument → x-axis values (week_day)

The second argument → y-axis values (temp)
```

```python
plt.plot(week_day,temp,'v:g',markersize=10)
plt.xlabel("weekday")
plt.ylabel("temperature")
plt.title("weekday vs temp")
plt.grid()
```

```python
days = [1,2,3,4,5,6,7] # days
max_t = [50,51,52,47,45,53,55] # max temp recor at day
min_t = [42,43,47,41,40,39,37]# min temp 
avg_t = [45,48,48,46,40,42,41]# avg temp
```

```python
plt.plot(days,max_t,label= "max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc = "upper left")
```

# Pie Chart

```python
exp_vals = [1400,600,300,400,100]
exp_labels = ["Home rent", "Food","Internet Bills","Car", "Other utilities"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(exp_vals,labels = exp_labels)
```

```python
plt.pie(exp_vals,labels = exp_labels,autopct = '%0.2f%%',radius = 1.5)
```

```python
plt.pie(exp_vals,labels = exp_labels,autopct = '%0.2f%%',radius = 1.5,explode = [0.04,0,0,0,0.08])
```
