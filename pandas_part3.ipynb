# concat :- joining of two  dataframes


```python
import pandas as pd

# Sample weather data for Indian cities
india_weather = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Kolkata"],
    "Temperature (°C)": [32, 30, 31],
    "Humidity (%)": [45, 70, 75],
    "Condition": ["Sunny", "Humid", "Rainy"]
})
india_weather
```

```python
import pandas as pd
us_weather = pd.DataFrame({
    "City": ["US", "tokyo", "South korea"],
    "Temperature (°C)": [22, 12, 15],
    "Humidity (%)": [56, 15, 18],
    "Condition": ["humid", "sunny", "cloudy"]
})
us_weather
```

```python
pd.concat([india_weather,us_weather] , ignore_index=True)
```

```python
pd.concat([india_weather,us_weather],axis=1)
```

```python
df = pd.concat([india_weather,us_weather],keys = ['india','us'])
```

```python
df.loc['us'].loc[1]
```

```python
df.loc['us',1]
```

```python
temperature_df = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore"],
    "temperature ": [32,42,53],
},index = [0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city" : ["Delhi","Mumbai"],
    "windspeed": [7,12],
},index = [1,0])
windspeed_df
```

```python
pd.concat([temperature_df,windspeed_df])
```

```python
pd.concat([temperature_df,windspeed_df],axis = 1)
```

# Merging

```python
df1= pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore"],
    "temperature ": [32,42,53],
})
df1
```

```python
df2= pd.DataFrame({
    "city" : ["Delhi","Mumbai","Banglore"],
    "humidity ": [45,70,43],
})
df2
```

```python
pd.merge(df1,df2, how = "inner")
```

```python
df1= pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore","Baltimore"],
    "temperature ": [32,42,53],
})
df1
```

```python
df2= pd.DataFrame({
    "city" : ["Delhi","Mumbai","Banglore"],
    "humidity ": [45,70,43],
})
df2
```

```python
pd.merge(df1,df2, how = "inner")
```

```python
df1= pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore","Baltimore"],
    "temperature ": [32,42,53,42],
})
df1
```

```python
df2= pd.DataFrame({
    "city" : ["Delhi","Mumbai","Banglore",'spain'],
    "humidity ": [45,70,43,56],
})
df2
```

```python
pd.merge(df1,df2, how = "left")
```

```python
pd.merge(df1,df2, how = "right")
```

```python
pd.merge(df1,df2, how = "outer")
```

```python
pd.merge(df1,df2, how = "outer",indicator = True)
```
