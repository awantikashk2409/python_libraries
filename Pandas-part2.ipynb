```python
space_org = {
    "Organisation": ["CNES","ISRO","NASA","JAXA","CNSA"],
    "Headquarters": ["Paris","Banglore","Washington","Tokyo","Beijing"],
    "Country_name": ["France","India","USA","Japan","China"],
    "Founded": [1958, 1969, 1975, 1992, 1993]  
}
```

```python
df = pd.DataFrame(space_org)
```

```python
df
```

```python
for i in df.itertuples():
    print(i,end='\n\n')
```

```python
for i in df.itertuples():
    print(i.Organisation,end='\n\n')
```

```python
# Write the query to show that all the headquarters present in  'france'
```

```python
for i in df.itertuples():
    if(i.Country_name == 'France'):
        print(i.Headquarters)
```

```python
for i in df.iterrows():
    print(i,end='\n\n')
```

```python
for i in df.iterrows():
    print(i[1]['Organisation'],end='\n\n')
```

```python
for i in df.items(): # (column name,data)
    print(i,end='\n\n')
```
